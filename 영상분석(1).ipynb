# ▩ Drawing Functions - Image Processing

### ▢ 그림 함수
* 선, 원, 사각형, 타원, 글자 등 그리기

### ㅇ Drawing Functions in OpenCV
- https://docs.opencv.org/4.x/dc/da5/tutorial_py_drawing_functions.html
- https://docs.opencv.org/4.x/d6/d6e/group__imgproc__draw.html

```python
### packages
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
### 영상 출력 함수
def fn_imshow(img, axis='off', fig_size=''):
    if fig_size!='': plt.figure(figsize=fig_size)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    if axis!='on': plt.axis('off')
    plt.show()
```

```python
### 영상 생성 - 세로:300, 가로:400
img = np.zeros((300,400,3), np.uint8) + 255
```

```python
### 영상 출력
fn_imshow(img, axis='on')
```

### ■ 선(line)
- 선의 시작 좌표와 끝 좌표를 사용

#### ㅇ line() - OpenCV
- cv.line(img, pt1, pt2, color[, thickness[, lineType]])
    * img: 영상
    * pt1: 선분의 첫 번째 점(시작 좌표) - ( )
    * pt2: 선분의 두 번째 점(끝 좌표) - ( )
    * color: 선 색상(BGR의 경우 튜플로 전달)
    - thickness: 선 두께(원과 같이 닫힌 도형에 -1 을 넘기면 도형을 채움. 기본 두께 = 1)
    - lineType: 라인의 유형(선종류. 기본값=8-connected. 곡선에 적합한 라인
    =cv2.LINE_AA)


```python
### 두꼐가 3픽셀인 파란색 대각선
cv2.line(img, (0,0), (400,300), (255,0,0), 3)

### 영상출력
fn_imshow(img, axis='on')
```

```python
### 중심을 지나는 수평선
cv2.line(img, (0,150), (400, 150), (0,255,0), 1)

### 중심을 지나는 수직선
cv2.line(img, (200,0), (200,300), (0,0,255), 1)

### 영상 출력
fn_imshow(img, axis='on')
```

### ■ 직사각형(rectangle)
- 직사각형의 왼쪽 위 모서리와 오른쪽 아래 모서리 좌표 사용

#### ㅇ rectangle() - OpenCV
- cv2.rectangle(img, pt1, pt2, color[, thickness[, lineType]])
    - img: 영상
    - pt1: 직사각형의 첫 번째 점(왼쪽 위 좌표)
    - pt2: 직사각형의 두 번째 점(오른쪽 아래 좌표)
    - color: 선 색상(BGR의 경우 튜플로 전달)
    - thickness: 선 두께(원과 같이 닫힌 도형에 -1 을 넘기면 도형을 채움. 기본 두께 = 1)
    - lineType: 라인의 유형(선종류. 기본값=8-connected. 곡선에 적합한 라인
    =cv2.LINE_AA)


```python
### 직사각형
cv2.rectangle(img, (0,0), (50,100), (0, 255, 0), 3)

fn_imshow(img, axis='on')
```

```python
### 직사각형
cv2.rectangle(img, (50, 0), (100,100), (0,255,0), -1)

fn_imshow(img, axis='on')
```

### ■ 원(circle)
- 중심 좌표와 반지름

#### ㅇ circle() - OpenCV
- cv.circle(img, center, radius, color[, thickness[, lineType]])
- img: 영상
- center: 원의 중심 좌표
- radius: 원의 반지름
- color: 원 색상
- thickness: 양수인 경우 원 외곽선의 두께
- lineType: 원 경계선의 유형


```python
### 원
cv2.circle(img, (350,50), 50, (0,0,255), -1)

fn_imshow(img, axis='on')
```

```python
### 점
cv2.circle(img, (350,50), 1, (0,0,0), -1)
### 영상 출력
fn_imshow(img, axis='on')
```

### ■ 타원(ellipse)
- 중심 위치(x,y)와 축 길이(주축 길이, 부축 길이)
- 회전 각도와 타원 호의 시작과 끝
### ㅇ ellipse() - OpenCV
- cv.ellipse(img, center, axes, angle, startAngle, endAngle, color[, thickness[, lineType]])
    - center: 중심 위치(x,y)
    - axes: 축 길이(주축 길이, 부축 길이)
    - angle: 회전 각도
    - startAngle, endAngle: 주축에서 시계 방향으로 측정된 타원 호의 시작과 끝

```python
### 타원
cv2.ellipse(img, (100,225), (100,50), 0, 0, 360, (255,0,0), -1)
### 영상 출력
fn_imshow(img, axis='on')
```

```python
### 타원
cv2.ellipse(img, (100,225), (100,50), 90, 0, 180, (255,0,255), -1)
### 영상 출력
fn_imshow(img, axis='on')
```

### ■  다각형(polygon)
- 꼭짓점의 좌표와 닫힘 여부 In [12]:
- 여러 줄을 그리는 데 사용할 수 있음
#### ㅇ polylines() - OpenCV
- cv2.polylines(img, pts, isClosed, color[, thickness[, lineType]])
    - pts: 다각형 곡선의 배열
    - isClosed: 다각형의 닫힘 여부


```python
### 꼭짓점의 좌표
pts = np.array([[50,50],[10,200],[250,250],[200,10]], np.int32)
### 다각형
cv2.polylines(img, [pts], True, (0,0,0), 2)
### 영상 출력
fn_imshow(img, axis='on')
```

### ■ 글자(text)
- 글자와 글자를 배치하려는 위치 좌표(왼쪽 하단 모서리)
- 글꼴의 유형과 크기를 지정
- lineType = cv.LINE_AA 권장
#### ㅇ putText() - OpenCV
- cv.putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType]])
    - img: 영상
    - text: 글자
    - org: 영상에 있는 문자열의 왼쪽 아래 모서리 좌표
    - fontFace: 글꼴 유형(Font type) 
    - fontScale: 글자 크기(글꼴별 기본 크기를 곱한 글꼴 배율 인수)
    - color: 텍스트 색상
    - thickness: 텍스트를 그리는 데 사용되는 선의 두께
    - lineType: 라인 유형

```python
### 글꼴 유형
font = cv2.FONT_HERSHEY_SIMPLEX # or 0
cv2.putText(img, 'OpenCV', (0,150), font, 2, (0,0,255), 2, cv2.LINE_AA)
### 영상 출력
fn_imshow(img, axis='on')
```
