```python
### Packages
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
### 출력 영상 크기
plt.rcParams["figure.figsize"] = (16,9) # 그림(figure)의 크기. (가로, 세로) 인치 단위
### 출력 영상에 한글 표시
plt.rcParams['font.family'] = "Gulim"
```

```python
### dtype=unit8
a = np.array(110, dtype='uint8')
b = np.array(200, dtype='uint8')
print('{0:08b}'.format(a))
print('{0:08b}'.format(b))
```

```python
### AND
c = np.bitwise_and(a , b)
print('{0:08b}'.format(c))
print(c)
```

```python
### OR
c = np.bitwise_or(a , b)
print('{0:08b}'.format(c))
print(c)
```

```python
### NOT
c = np.invert(a)
print('{0:08b}'.format(c))
print(c)
```

```python
### 영상 읽기
img_gray = cv2.imread(r'lena.png', 0)
img_gray.shape
```

```python
### Mask
img_mask = np.zeros((512, 512), dtype='uint8')
img_mask[128:385, 128:385] = 255
img_mask.shape
```

```python
### 논리연산
img_and = cv2.bitwise_and(img_gray, img_mask)
img_or = cv2.bitwise_or(img_gray, img_mask)
img_xor = cv2.bitwise_xor(img_gray, img_mask)
img_not = cv2.bitwise_not(img_gray)
```

```python

### 영상 출력
titles = ["원래 영상", "mask", "bitwise_and", "bitwise_or", "bitwise_xor", "bitwise_not"]
images = [img_gray, img_mask, img_and, img_or, img_xor, img_not]
for i in range(len(images)):
    img_rgb = cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB)
    plt.subplot(2, 3, i+1)
    plt.imshow(img_rgb)
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

```python
### 영상 읽기
img_raw = cv2.imread(r'lena.png')
img_raw.shape
```

```python

### Mask
img_mask = np.zeros((512, 512, 3), dtype='uint8')
img_mask[128:385, 128:385, :] = 255
img_mask.shape
```

```python
### 영상 출력
titles = ["원래 영상", "mask", "bitwise_and", "bitwise_or", "bitwise_xor", "bitwise_not"]
images = [img_raw, img_mask, img_and, img_or, img_xor, img_not]
for i in range(len(images)):
    img_rgb = cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB)
    plt.subplot(2, 3, i+1)
    plt.imshow(img_rgb)
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

```python
### 입력 영상
img_in = np.zeros((25, 25), dtype='uint8')
img_in[8:17, 8:17] = 255
img_in[12, 12] = 0
```

```python
### 커널
kernel_1 = np.ones((3, 3)) / (3*3)
kernel_2 = np.array([[1, 2, 1],
    [2, 4, 2],
    [1, 2, 1]]) / 16
kernel_3 = np.array([[2, 3, 2],
    [3, 5, 3],
    [2, 3, 2]]) / 25
```

```python
### 저주파 통과 필터
img_1 = cv2.filter2D(img_in, -1, kernel_1)
img_2 = cv2.filter2D(img_in, -1, kernel_2)
img_3 = cv2.filter2D(img_in, -1, kernel_3)
```

```python
### 영상 출력
titles = ["원래 영상", "kernel 1", "kernel 2", "kernel 3"]
images = [img_in, img_1, img_2, img_3]
for i in range(len(images)):
    img_rgb = cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB)
    plt.subplot(1,4,i+1), plt.imshow(img_rgb)
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

```python
### 커널의 크기
kernel_1 = np.ones((3, 3)) / (3**2)
kernel_2 = np.ones((5, 5)) / (5**2)
kernel_3 = np.ones((7, 7)) / (7**2)
```

```python
### 저주파 통과 필터 - Averaging: cv2.blur()
img_1 = cv2.filter2D(img_lena, -1, kernel_1) # cv2.blur(img_lena, (3, 3))
img_2 = cv2.filter2D(img_lena, -1, kernel_2) # cv2.blur(img_lena, (5, 5))
img_3 = cv2.filter2D(img_lena, -1, kernel_3) # cv2.blur(img_lena, (7, 7))
```

```python
### 영상 출력(3)
x1, x2, y1, y2 = [300, 331, 70, 101]
titles = ["원래 영상", "Size=3", "Size=5", "Size=7"]
images = [img_lena, img_1, img_2, img_3]
for i in range(len(images)):
    img_rgb = cv2.cvtColor(images[i][x1:x2, y1:y2, :], cv2.COLOR_BGR2RGB)
    plt.subplot(1, 4, i+1), plt.imshow(img_rgb)
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

```python
### 가우시안 필터(Gaussian filter)
img_1 = cv2.GaussianBlur(img_lena, (3, 3), 0)
img_2 = cv2.GaussianBlur(img_lena, (5, 5), 0)
img_3 = cv2.GaussianBlur(img_lena, (7, 7), 0)
```
